## Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
sns.set_style('whitegrid')
```

### Labeling plot features
#### The functional method

```python
import matplotlib.pyplot as plt

x = range(1,10)
y = [1, 2, 3, 4, 0, 5, 4, 3, 2]  # Adjusted to 9 elements

plt.bar(x, y)
plt.xlabel('Your x-axis label')
plt.ylabel('Your y-axis label')
plt.show()
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle','moterbike','car','van','stroller']
plt.pie(z,labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/cars.csv'

# Read the CSV file
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_ticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_xlabel('car names')   
ax.set_ylabel('miles/gal')            
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_ticklabels(cars.car_names,rotation=60,fontsize='medium')
ax.set_xlabel('car names')   
ax.set_ylabel('miles/gal')  
```
