```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
```

```python
import pandas as pd

data = {
    'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'micheal'],
    'age': [20, 22, 20, 21, 24, 23],
    'gender': ['male', 'male', 'female', 'female', 'male', 'male'],  # Corrected
    'rank': [2, 1, 4, 5, 3, 7]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

### Matplotlib's Bar Chart

```python
plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('camparing age')
plt.show()
```

```python
plot = sns.barplot(data=df,x='names',y='age')
plot.set_title('comparing ages')
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('camparing age')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(data=df,x='names',y='age')
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'],labels=df['names'])
plt.title('age comparsion')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:4]
plt.pie(df['age'],labels=df['names'],colors=colors)
plt.show()
```
